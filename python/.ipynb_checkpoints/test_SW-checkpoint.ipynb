```python
path = "../../../data/dna/"
file_res = "results.txt"
file_res_shuffled = "results_shuffled.txt"
```

```python
with open(path + file_res, 'r') as file:
    results = np.array([float(item) for item in ''.join(file.readlines()).split('\n')[:-1]])
results[-10:]
```

```python
with open(path + file_res_shuffled, 'r') as file:
    results_shuffled = np.array([float(item) for item in ''.join(file.readlines()).split('\n')[:-1]])
results_shuffled[-10:]
```

```python
plt.figure(figsize=(14, 8))

limit = 50

x = np.array([i for i in range(results.shape[0])])[:limit]

plt.plot(x, results[:limit], label='origin')
plt.plot(x, results_shuffled[:limit], label='shuffled')

# plt.ylim(-200)

plt.legend()

plt.show()
```

```python
path = "C:/Users/kshch/Projects/diploma/source/data/dna/"
file_name = "exon_intron_forward_200.fasta"
file_name_sh = "exon_intron_forward_200_shuffled.fasta"
```

```python
with open(path + file_name, 'r') as file:
    lines = file.readlines()
    lines = ''.join(''.join(lines).split('\n'))
```

```python
lines
```

```python
lines = list(lines)
np.random.shuffle(lines)
lines = ''.join(lines)
lines
```

```python
with open(path + file_name_sh, 'a') as file:
    index1 = 0
    index2 = 100
    while index2 < len(lines):
        file.write(lines[index1:index2])
        file.write('\n')
        index1 += 100
        index2 += 100
```
